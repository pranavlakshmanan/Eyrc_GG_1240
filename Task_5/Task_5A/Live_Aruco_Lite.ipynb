```python
import cv2 as cv
import numpy as np


def undistort_image(img, npz_file):
    data = np.load(npz_file)
    mtx = data["camMatrix"]
    dist = data["distCoef"]
    h, w = img.shape[:2]
    newcameramtx, roi = cv.getOptimalNewCameraMatrix(mtx, dist, (w, h), 1, (w, h))
    dst = cv.undistort(img, mtx, dist, None, newcameramtx)
    x, y, w, h = roi
    dst = dst[y : y + h, x : x + w]

    return dst
```

```python
# Initialize the video capture object
try:
    cap = cv.VideoCapture("/dev/video2")
    if not cap.isOpened():
        print("Could not open /dev/video2")
        raise Exception("Could not open /dev/video2")
except:
    cap = cv.VideoCapture("/dev/video3")
    if not cap.isOpened():
        print("Could not open /dev/video3 either. Please check your device.")
        try:
            cap = cv.VideoCapture("/dev/video0")
            if not cap.isOpened():
                print("Could not open /dev/video0")
                raise Exception("Could not open /dev/video0")
        except:
            cap = cv.VideoCapture("/dev/video1")
            if not cap.isOpened():
                print("Could not open /dev/video1 either. Please check your device.")
desired_width = 1920
desired_height = 1080
cap.set(cv.CAP_PROP_FRAME_WIDTH, desired_width)
cap.set(cv.CAP_PROP_FRAME_HEIGHT, desired_height)


def detect_aruco(frame):
    gray = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)
    # gray = cv.GaussianBlur(gray, (3, 3), 3)
    dictionary = cv.aruco.getPredefinedDictionary(cv.aruco.DICT_4X4_1000)
    parameters = cv.aruco.DetectorParameters()
    corners, ids, rejectedImgPoints = cv.aruco.detectMarkers(
        gray, dictionary, parameters=parameters
    )
    centers = []
    cv.aruco.drawDetectedMarkers(frame, corners, ids)
    for i in range(len(ids)):
        center = np.mean(corners[i], axis=1, dtype=int)
        centers.append(center)
        markerID = ids[i][0]
        markerCorners = corners[i][0]
        centerX = int((markerCorners[0, 0] + markerCorners[2, 0]) / 2.0)
        centerY = int((markerCorners[0, 1] + markerCorners[2, 1]) / 2.0)

        cv.putText(
            frame,
            str(markerID),
            (centerX, centerY),
            cv.FONT_HERSHEY_SIMPLEX,
            1.0,
            (0, 255, 0),
            2,
        )

    return ids, centers


while True:
    ret, frame = cap.read()
    # frame = cv.rotate(frame, cv.ROTATE_90_COUNTERCLOCKWISE)[450:1600, 0:1080]
    frame = cv.rotate(frame, cv.ROTATE_90_COUNTERCLOCKWISE)[450:1500,0:1080]
    frame = undistort_image(
        frame,
        "/mnt/Storage/Projects/E-YRC/EYRC_2023/Task_5/Task_5A/MultiMatrix.npz",
    )
    ids, centers = detect_aruco(frame)
    print("len",len(ids))
    cv.imshow("Image", cv.resize(frame, (900, 900)))
    if cv.waitKey(1) & 0xFF == ord("q"):
        break

cap.release()
cv.destroyAllWindows()
```
