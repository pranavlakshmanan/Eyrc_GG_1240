```python
class ImageThresholdApp:
    def __init__(self, root, image_path):
        self.root = root
        self.root.title("Image Thresholding App")

        # Load the image
        self.image = cv.imread(image_path, cv.IMREAD_GRAYSCALE)

        # Initialize default values for upper and lower thresholds
        self.lower_threshold = 0
        self.upper_threshold = 255

        # Create a label to display the original image
        self.original_label = ttk.Label(self.root, text="Original Image")
        self.original_label.grid(row=0, column=0, padx=10, pady=10)

        # Display the original image
        self.display_image(self.image)

        # Create sliders for upper and lower thresholds
        self.lower_slider = self.create_slider("Lower Threshold", 0, 255, self.lower_threshold, 1, 0)
        self.upper_slider = self.create_slider("Upper Threshold", 0, 255, self.upper_threshold, 1, 255)

        # Create a button to apply thresholding
        apply_button = ttk.Button(self.root, text="Apply Thresholding", command=self.apply_thresholding)
        apply_button.grid(row=2, column=0, pady=10)

        cv.waitKey(0)
        cv.destroyAllWindows()

        

    def create_slider(self, label_text, from_, to, default_value, resolution, initial_position):
        ttk.Label(self.root, text=label_text).grid(row=1, column=0, padx=10, pady=5)
        slider = ttk.Scale(self.root, from_=from_, to=to, orient=tk.HORIZONTAL, length=200,
                           variable=tk.DoubleVar(value=default_value), command=self.update_slider_value)
        slider.set(initial_position)
        slider.grid(row=1, column=1, padx=10, pady=5)
        return slider

    def update_slider_value(self, value):
        pass

    def display_image(self, image):
        cv.imshow("image",image)
        # height, width = image.shape[:2]
        # image = cv.resize(image, (int(width / 2), int(height / 2)))
        # image = cv.cvtColor(image, cv.COLOR_BGR2RGB)
        # img = tk.PhotoImage(width=image.shape[1], height=image.shape[0])
        # img.put("#%02x%02x%02x" % (255, 255, 255), (0, 0, image.shape[1], image.shape[0]))
        # img.put(image.flatten(), (0, 0, image.shape[1], image.shape[0], 0, 0))
        # tk.Label(self.root, image=img).grid(row=0, column=1, padx=10, pady=10)

    def apply_thresholding(self):
        # Apply thresholding to the image
        _, thresholded_image = cv.threshold(self.image, self.lower_threshold, self.upper_threshold, cv.THRESH_BINARY)

        # Display the thresholded image
        self.display_image(thresholded_image)

    def run(self):
        self.root.mainloop()
```

```python
event_list = []


def event_identification(arena):
    arena_org = arena.copy()
    image_cnts = 0
    # l_tresh, u_tresh = slide_tresh()
    # print(l_tresh, u_tresh)
    arena_gray = cv.cvtColor(arena, cv.COLOR_BGR2GRAY)
    # cv.imshow("gray", arena_gray)
    arena_gray = cv.GaussianBlur(arena_gray, (5, 5), 1)
    root = tk.Tk()
    app = ImageThresholdApp(root, arena)
    app.run()
    thresh = app.apply_thresholding()
    # ret, thresh = cv.threshold(arena_gray, l_tresh, u_tresh, cv.THRESH_BINARY)
    # ret, thresh = cv.threshold(arena_gray, 80, 255, cv.THRESH_BINARY)

    cv.imshow("thresh", thresh)
    contours, hierarchy = cv.findContours(thresh, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)
    for cnt in contours:
        approx = cv.approxPolyDP(cnt, 0.01 * cv.arcLength(cnt, True), True)
        if len(approx) == 4:
            x1, y1, w, h = cv.boundingRect(cnt)
            ratio = float(w) / h
            area = w * h
            if ratio >= 0.9 and ratio <= 1.1:
                # cropped_image = arena_org[
                #     y1 + border_width : y1 + 50 - border_width,
                #     x1 + border_width : x1 + 50 - border_width,
                # ]
                image_cnts += 1
                event_list.append([x1, y1, w, h])
    return event_list, image_cnts
```

```python
cap = cv.VideoCapture("/dev/video0")
desired_width = 1980
desired_height = 1080
cap.set(cv.CAP_PROP_FRAME_WIDTH, desired_width)
cap.set(cv.CAP_PROP_FRAME_HEIGHT, desired_height)


while True:
    # Capture the video frame
    # by frame
    ret, frame = cap.read()
    event_list, image_cnts = event_identification(frame)
    print("event_list", event_list)
    print("image_cnts", image_cnts)
    for i in range(0, image_cnts):
        frame = cv.rectangle(
            frame,
            (event_list[i][0], event_list[i][1]),
            (event_list[i][0] + event_list[i][2], event_list[i][1] + event_list[i][3]),
            (0, 255, 0),
            3,
        )
    # frame = cv.resize(frame, (1800,1000))
    cv.imshow("Image", frame)
    # cv.waitKey(0)
    # cv.destroyAllWindows()
    # Display the resulting frame

    # cv.imshow('frame', frame)

    # the 'q' button is set as the
    # quitting button you may use any
    # desired button of your choice
    if cv.waitKey(1) & 0xFF == ord("q"):
        break

cap.release()
cv.destroyAllWindows()
```
